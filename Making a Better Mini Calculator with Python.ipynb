# Making a Calculator with Python
A difference from using C-sharp, is that python's version of a switch case is a match/case. So, instead of splitting the switch cases, a single block of code is suffcient (and at times more efficient) for a match/case. I have an example below of what code you would use if you have a version of python 3.11 and above (3.9 and below will not run the mactch choice example). But to make the calculcator universal, I also did a version that makes use of python's functions ( better known as: def).

def calculator():
    result = 0
    
    while True:
        print(f"\nCurrent Value of the Calcultor: {result}")
        choice = input("Enter what operation you'd like to do from these: (+, -, *, /, -/+, %, ^, AC, or 'ex' to exit).").strip()
        
        match choice:
            #First four match/case's is the basic addition, subtraction, multiplication, and division.
            case '+':
                num = float(input("Enter the number you'd like to add: "))
                result += num
                
            case '-':
                num = float(input("Enter the number you'd like to subtract: "))
                result -= num
                
            case '*':
                num = float(input("Enter the number you'd like to multiply: "))
                result *= num
                
            case '/':
                num = float(input("Enter the number you'd like to divide: "))
            #To handle if someone tries to divide by zero then I'll make a conditional if-else statement.
                if num != 0
                result /= num
                else:
                    print("You cannot divide by zero.")
                    
            #This would be (in C-Sharp) the second switch case.
            case '-/+':
                #Toggle positive or negative sign
                result = -result
                
            case '%':
                #Percentage of a number.
                result = result / 100
                
            case '^':
                #Exponent
                num = float(input("Enter the exponent for the number: "))
                result = result ** num
                
            case 'AC':
                #All Clear.
                result = 0
                
            case 'ex':
                print("You've exited the calculator app- Goodbye!")
                break
                
            case _:
                print("This is not an action that can be handled in this small app. Please try again.")
                
    if __name__ == "__main__":
        calculator ()
        
###The code above does works as long as your version is 3.11 but is the version is 3.9 ar below then you'll need to upgrade- but this is how a match/case can be utilized in python just as a switch case is used in C-Sharp.

```python
def calculator():
    print("Hi! Welcome to the python version of my C Sharp: My calculator, but better App! You can type 'exit' to quit.")
    
    def add(a, b): return a + b
    def subtract(a, b): return a - b
    def multiply(a, b): return a * b
    def divide(a, b): return a / b if b != 0 else "You can't divide by Zero"
    def toggle_sign(a): return -a
    def percent(a, b): return (a / b) * 100 if b != 0 else "You can't divide by Zero"
    def exponent(a, b): return a ** b
    def clear(): return "Clear!"
    
    operations = {
        '+' : add,
        '-': subtract,
        '*' : multiply,
        '/' : divide,
        '+/-' : toggle_sign,
        '%': percent,
        '^' : exponent,
        'AC' : clear,
    }
    
    while True:
        choose = input("\nChoose what operation you'd like to use (+, -, *, /, -/+, %, ^, AC,) or 'exit': ")
        
        if choose.lower() == "exit":
            print("Thank's for stopping by!")
            break
        
        if choose not in operations:
            print("Please follow the prompts of the program to get a valid response.")
            continue
            
        #Handling single operations:
        if choose in ["+/-", "AC"]:
            if choose == "+/-":
                a = float(input("Enter a number here: "))
                result = operations[choose](a)
            else: 
                #This is for the All Clear
                result = operations[choose]()
                print(f"Final Result: {result}")
                continue
                
        #Everything else needs two numbers.
        try:
            a = float(input("Enter the first number here: "))
            b = float(input("Enter the second number here: "))
            result = operations[choose](a, b)
            print(f"Final Result: {result}")
        except ValueError:
                print("Please follow the prompts of the program to get a valid response.")
calculator()
```
