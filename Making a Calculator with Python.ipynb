# Making a Calculator with Python
A difference from using C-sharp, is that python's version of a switch case is a match/case. Instead of splitting the switch cases, a single block of code is suffcient (and at times more efficient) for a match/case. So this calcultor has a single match/case as opposed to when I made a C-sharp calculator which contained two switch cases.

```python
def calculator():
    result = 0
    
    while True:
        print(f"\nCurrent Value of the Calcultor: {result}")
        choice = input("Enter what operation you'd like to do from these: (+, -, *, /, -/+, %, ^, AC, or 'ex' to exit).").strip()
        
        match choice:
            #First four match/case's is the basic addition, subtraction, multiplication, and division.
            case '+':
                num = float(input("Enter the number you'd like to add: "))
                result += num
                
            case '-':
                num = float(input("Enter the number you'd like to subtract: "))
                result -= num
                
            case '*':
                num = float(input("Enter the number you'd like to multiply: "))
                result *= num
                
            case '/':
                num = float(input("Enter the number you'd like to divide: "))
            #To handle if someone tries to divide by zero then I'll make a conditional if-else statement.
                if num != 0
                result /= num
                else:
                    print("You cannot divide by zero.")
                    
            #This would be (in C-Sharp) the second switch case.
            case '-/+':
                #Toggle positive or negative sign
                result = -result
                
            case '%':
                #Percentage of a number.
                result = result / 100
                
            case '^':
                #Exponent
                num = float(input("Enter the exponent for the number: "))
                result = result ** num
                
            case 'AC':
                #All Clear.
                result = 0
                
            case 'ex':
                print("You've exited the calculator app- Goodbye!")
                break
                
            case _:
                print("This is not an action that can be handled in this small app. Please try again.")
                
    if __name__ == "__main__":
        calculator ()
        
#This works as long as your version is 3.11 but is the version is 3.9 ar below then you'll need to upgrade.
```

```python
def calculator():
    result = 0
    
    while True:
        print(f"\nCurrent Value of the Calcultor: {result}")
        choice = input("Enter what operation you'd like to do from these (+, -, *, /, -/+, %, ^, AC, or 'ex' to exit): ").strip()
        
        if choice == 'ex':
            print ("You've exited the calculator app- Goodbye!")
            break
            
            if choice in ['+', '-', '*', '/']:
                num = float(input("Please enter a number: "))
                
                operations = {
                    '+' : lambda r, n: r + n,
                    '-' : lambda r, n: r - n,
                    '*' : lambda r, n: r * n,
                    '/' : lambda r, n: r / n if n != 0 else float('inf')
                }
                result = operation[choice](result, num)
                
            else: 
                special = {
                    '+/-' : lambda r: -r,
                    '%' :   lambda r: r / 100,
                    '^' :   lambda r: r ** float(input("Please enter the exponent: ")),
                    'AC' :  lambda r: 0
                }
                
                if choice in special:
                    result = special[choice](result)
                else:
                    print("This is not an action that can be handled in this small app. Please try again.")
                
    if __name__ == "__main__":
        calculator ()
#This is a 'use dictionary switch', and it's backwards compatible so it works in all versions and simulates the switch case.
```
