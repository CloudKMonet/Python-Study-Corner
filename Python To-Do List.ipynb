# Making a To-Do List in Python

The focus is on the CRUD operations. I have also done a C# app version of this!

```python
#The first thing is to establish that each to-do is it's own array

todos = []

#Each task has an ID#, a title, and a true or false to indicate if its completed.

#Define the CRUD functions: Create, Read, Update, and Delete
def create_todo(title):
    todo_id = len(todos) + 1
    todos.append({"id": todo_id, "title": title, "completed": False})
    print(f"To-Do '{title}' added with ID '{todo_id}'.")

def list_todo():
    if not todos:
        print("No To-Do's, written... yet.... :D")
        return
    for todo in todos:
        status = "✅" if todo["completed"] else "❌"
        print(f"{todo['id']}: {todo['title']} [{status}]")
        
def update_todo(todo_id, new_title=None, completed=None):
    for todo in todos:
        if todo["id"] == todo_id:
            if new_title:
                todo["title"] = new_title
                if completed is not None:
                    todo["completed"] = completed
                    print(f"To-Do {todo_id} has been updated.")
                    return
    print(f"To-Do {todo_id} is not found.")

def delete_todo(todo_id):
    global todos
    todos = [todo for todo in todos if todo["id"] != todo_id]
    print(f"To-Do {todo_id} has been deleted.")
```

```python
def main():
    while True:
        print("----- Welcome! To-Do List Options -----")
        print("1. Add To-Do")
        print("2. View To-Do")
        print("3. Update To-Do")
        print("4. Delete To-Do")
        print("5. Exit")
        
        choice = input("Please choose an option to perform: ")
        
        if choice == "1":
            title = input("Enter your To-Do here: ")
            create_todo(title)
        elif choice == "2":
            list_todo()
        elif choice == "3":
            todo_id = int(input("Enter your To-Do list ID to update: "))
            new_title = input("Enter new title (leave blank to skip this step): ")
            completed_input = input("Mark as completed? (yes/no/skip): ").lower()
            completed = None
            if completed_input == "yes":
                completed = True
            elif completed_input == "no":
                completed = False
            update_todo(todo_id, new_title or None, completed)
        elif choice == "4":
            todo_id = int(input("Enter the To-Do ID to delete the To-Do: "))
            delete_todo(todo_id)
        elif choice == "5":
            print("Bye!")
            break
        else:
            print("That response is invalid. Please try again.")
if __name__ == "__main__":
    main()
```
